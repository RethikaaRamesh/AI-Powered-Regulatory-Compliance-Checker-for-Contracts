# Clause Identification Program

## Objective:
This program identifies whether important regulatory clauses
such as GDPR and HIPAA are present in a contract document.

## How it works:
1. Store the contract text in a variable.
2. Define a list of required regulatory clauses.
3. Check if each clause exists in the contract.
4. Display the clauses that are found and those that are missing.

## Use case:
This module helps in automatic contract review by detecting
whether required compliance clauses are included.

```python

# Clause Identification Program

# Step 1: Store contract text
contract_text = """
This agreement follows GDPR regulations and ensures data protection.
The company shall report any breach within 72 hours.
"""

# Step 2: Define required clauses
required_clauses = ["GDPR", "HIPAA", "data protection"]

# Step 3: Create lists to store found and missing clauses
found_clauses = []
missing_clauses = []

# Step 4: Check each clause in contract
for clause in required_clauses:
    if clause.lower() in contract_text.lower():
        found_clauses.append(clause)
    else:
        missing_clauses.append(clause)

# Step 5: Display results
print("Contract Text:")
print(contract_text)

print("\nFound Clauses:")
print(found_clauses)

print("\nMissing Clauses:")
print(missing_clauses)

# Step 6: Simple compliance decision
if missing_clauses:
    print("\n‚ö†Ô∏è Compliance Risk: Missing clauses detected.")
else:
    print("\n‚úÖ Contract is compliant with required regulations.")
```

# Risk Analysis Program for Contract Compliance

## Objective:
This program analyzes a contract to determine compliance risk
by checking whether mandatory regulatory clauses are missing.

## How it works:
1. Take the list of clauses found in a contract.
2. Compare it with a list of required regulatory clauses.
3. Identify which clauses are missing.
4. Assign a risk level (High or Low) based on missing clauses.

## Use case:
This module helps in automatically assessing legal and regulatory
risks in contracts by flagging missing compliance clauses.

```python
# Risk Analysis Program

# Step 1: Clauses found from previous module (Clause Identification)
found_clauses = ["GDPR", "data protection"]

# Step 2: Clauses required for compliance
required_clauses = ["GDPR", "HIPAA", "data protection"]

# Step 3: List to store missing clauses
missing_clauses = []

# Step 4: Compare found and required clauses
for clause in required_clauses:
    if clause not in found_clauses:
        missing_clauses.append(clause)

# Step 5: Display missing clauses
print("Found Clauses:", found_clauses)
print("Required Clauses:", required_clauses)
print("Missing Clauses:", missing_clauses)

# Step 6: Risk level assessment
if len(missing_clauses) == 0:
    risk_level = "LOW RISK (Compliant Contract)"
elif len(missing_clauses) == 1:
    risk_level = "MEDIUM RISK (One clause missing)"
else:
    risk_level = "HIGH RISK (Multiple clauses missing)"

# Step 7: Show risk result
print("\nRisk Analysis Result:")
print("Risk Level:", risk_level)
```

# Regulatory Update Tracking Program

## Objective:
This program simulates tracking regulatory updates (such as GDPR and HIPAA)
and checking whether a contract needs to be updated based on new rules.

## How it works:
1. Store the latest regulatory updates in a dictionary.
2. Store the regulations already covered by the contract.
3. Compare both to find which regulations have new updates.
4. Display alerts if the contract requires review.

## Use case:
This module helps in continuously monitoring legal changes and ensuring
that active contracts stay compliant with updated regulations.

```python
# Regulatory Update Tracking Program

# Step 1: Store latest regulatory updates (simulated data)
regulatory_updates = {
    "GDPR": "New breach notification rule - Jan 2026",
    "HIPAA": "Mandatory encryption update - Feb 2026",
    "PCI-DSS": "Card data protection update - Mar 2026"
}

# Step 2: Regulations already covered in the contract
contract_regulations = ["GDPR", "PCI-DSS"]

# Step 3: Check for missing or new regulatory updates
updates_needed = []

for regulation in regulatory_updates:
    if regulation not in contract_regulations:
        updates_needed.append(regulation)

# Step 4: Display current updates
print("Latest Regulatory Updates:")
for law, update in regulatory_updates.items():
    print(f"{law} -> {update}")

# Step 5: Display which updates require contract changes
print("\nContract Currently Covers:", contract_regulations)
print("Regulations Requiring Review:", updates_needed)

# Step 6: Decision message
if updates_needed:
    print("\n‚ö†Ô∏è Contract requires updates for the above regulations.")
else:
    print("\n‚úÖ Contract is up-to-date with all regulations.")
```

# Contract Modification Program

## Objective:
This program suggests modifications to a contract when required
regulatory clauses are missing. It automatically generates
recommended clauses to make the contract compliant.

## How it works:
1. Take the list of missing clauses identified during risk analysis.
2. For each missing clause, generate a suggested legal statement.
3. Display the suggested clauses to be added to the contract.

## Use case:
This module helps in automatically updating contracts by providing
recommended clauses based on regulatory requirements.

```python
# Contract Modification Program

# Step 1: Missing clauses identified from Risk Analysis module
missing_clauses = ["HIPAA", "PCI-DSS"]

# Step 2: Predefined clause templates (simulated AI suggestions)
clause_templates = {
    "HIPAA": "This contract shall comply with HIPAA regulations for protecting patient health information.",
    "GDPR": "This contract shall ensure full compliance with GDPR data privacy and protection standards.",
    "PCI-DSS": "This contract shall follow PCI-DSS standards for secure handling of cardholder data."
}

# Step 3: Generate suggested modifications
suggested_clauses = []

for clause in missing_clauses:
    if clause in clause_templates:
        suggested_clauses.append(clause_templates[clause])

# Step 4: Display suggested contract modifications
print("Missing Clauses:", missing_clauses)
print("\nSuggested Contract Modifications:\n")

for i, clause in enumerate(suggested_clauses, start=1):
    print(f"{i}. {clause}")
```

# Integrated Python + SQL Compliance Checker System

## Objective
This program implements a prototype of an AI-powered regulatory
compliance checker using Python and SQL. The objective is to analyze
contract text, identify missing regulatory clauses, assess compliance
risk, store results in a database, and generate notifications for
compliance issues.

---

## System Description
The system uses Python for processing contract text and decision-making,
and SQLite (SQL) for storing contracts, clauses, risks, regulatory
updates, and notifications. The integration of Python and SQL allows the
program to simulate an intelligent contract compliance system.

---

## Working Principle
The program follows these steps:

1. **Database Connection and Table Creation**  
   The system connects to an SQLite database and creates tables to store:
   - Contracts  
   - Regulatory clauses  
   - Compliance risks  
   - Regulatory updates  
   - Notifications  

2. **Data Insertion**  
   Sample contract data and regulatory clauses are inserted into the
   database for demonstration purposes.

3. **Clause Identification**  
   The contract text is retrieved from the database and checked against
   a list of required clauses (GDPR, HIPAA, PCI-DSS).  
   Clauses found in the contract are marked as present, and missing
   clauses are identified.

4. **Risk Analysis**  
   The number of missing clauses is used to determine the risk level:
   - Low Risk: No missing clauses  
   - Medium Risk: One missing clause  
   - High Risk: More than one missing clause  

   The missing clauses and risk level are stored in the database.

5. **Regulatory Update Tracking**  
   The system retrieves and displays regulatory updates stored in the
   database to simulate monitoring of changes in laws and regulations.

6. **Contract Modification Suggestions**  
   For each missing clause, the system generates a suggested contract
   clause using predefined templates. These suggestions help in making
   the contract compliant with regulatory requirements.

7. **Notification Generation**  
   A compliance alert message is created based on the risk level and
   missing clauses. The alert is stored in the database and displayed as
   a simulated email or Slack notification.

8. **Final Compliance Report**  
   The system displays:
   - Compliance risks stored in the database  
   - Notification logs  

   This provides a complete summary of the contract‚Äôs compliance status.

---

## Database Structure
The following tables are used in the system:

- **contracts** ‚Äì stores contract text  
- **clauses** ‚Äì stores regulatory clause names  
- **compliance_risks** ‚Äì stores missing clauses and risk level  
- **regulatory_updates** ‚Äì stores law update information  
- **notifications** ‚Äì stores compliance alert messages  

This structure supports clause detection, risk analysis, regulatory
tracking, contract modification, and notification generation.

---

## Technologies Used
- Python for program logic and text processing  
- SQLite (SQL) for database storage and queries  
- String matching for clause detection  
- Database integration using the sqlite3 library  

---

## Conclusion
This integrated Python and SQL system demonstrates how contract
compliance checking can be automated. The system reduces manual review
effort by automatically identifying missing regulatory clauses,
assessing compliance risk, and generating alerts.

The system can be extended in the future using advanced Natural Language
Processing (NLP) techniques and Large Language Models (LLMs) such as GPT
or LLaMA for more accurate clause extraction and legal interpretation.

```python
import sqlite3

# =========================
# CONNECT TO DATABASE
# =========================
conn = sqlite3.connect("compliance.db")
cur = conn.cursor()

# =========================
# CREATE TABLES
# =========================
cur.execute("""
CREATE TABLE IF NOT EXISTS contracts (
    contract_id INTEGER PRIMARY KEY,
    contract_text TEXT
)
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS clauses (
    clause_id INTEGER PRIMARY KEY,
    clause_name TEXT
)
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS compliance_risks (
    risk_id INTEGER PRIMARY KEY,
    contract_id INTEGER,
    missing_clause TEXT,
    risk_level TEXT
)
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS regulatory_updates (
    update_id INTEGER PRIMARY KEY,
    regulation TEXT,
    update_info TEXT
)
""")

cur.execute("""
CREATE TABLE IF NOT EXISTS notifications (
    notification_id INTEGER PRIMARY KEY,
    contract_id INTEGER,
    message TEXT
)
""")

print("‚úÖ Tables created")

# =========================
# INSERT SAMPLE DATA
# =========================
cur.execute("DELETE FROM contracts")
cur.execute("DELETE FROM clauses")
cur.execute("DELETE FROM compliance_risks")
cur.execute("DELETE FROM regulatory_updates")
cur.execute("DELETE FROM notifications")

cur.execute("INSERT INTO contracts VALUES (1, 'This agreement follows GDPR rules and ensures data protection.')")

clauses = [(1, "GDPR"), (2, "HIPAA"), (3, "PCI-DSS")]
cur.executemany("INSERT INTO clauses VALUES (?,?)", clauses)

updates = [
    (1, "GDPR", "New breach reporting rule - Jan 2026"),
    (2, "HIPAA", "Mandatory encryption update - Feb 2026")
]
cur.executemany("INSERT INTO regulatory_updates VALUES (?,?,?)", updates)

conn.commit()
print("‚úÖ Sample data inserted")

# =========================
# CLAUSE IDENTIFICATION
# =========================
contract = cur.execute("SELECT contract_text FROM contracts WHERE contract_id=1").fetchone()[0]
required_clauses = [row[0] for row in cur.execute("SELECT clause_name FROM clauses")]

found_clauses = []
missing_clauses = []

for clause in required_clauses:
    if clause.lower() in contract.lower():
        found_clauses.append(clause)
    else:
        missing_clauses.append(clause)

print("\nüìÑ Contract Text:", contract)
print("‚úÖ Found Clauses:", found_clauses)
print("‚ùå Missing Clauses:", missing_clauses)

# =========================
# RISK ANALYSIS
# =========================
if len(missing_clauses) == 0:
    risk_level = "LOW"
elif len(missing_clauses) == 1:
    risk_level = "MEDIUM"
else:
    risk_level = "HIGH"

print("\n‚ö†Ô∏è Risk Level:", risk_level)

# Store risk in database
for clause in missing_clauses:
    cur.execute("INSERT INTO compliance_risks VALUES (NULL, ?, ?, ?)", (1, clause, risk_level))

conn.commit()

# =========================
# REGULATORY UPDATE CHECK
# =========================
print("\nüì¢ Regulatory Updates:")
for row in cur.execute("SELECT * FROM regulatory_updates"):
    print(row)

# =========================
# CONTRACT MODIFICATION
# =========================
clause_templates = {
    "HIPAA": "This contract shall comply with HIPAA patient data protection laws.",
    "PCI-DSS": "This contract shall comply with PCI-DSS card data security standards."
}

print("\n‚úçÔ∏è Suggested Contract Modifications:")
for clause in missing_clauses:
    if clause in clause_templates:
        print("-", clause_templates[clause])

# =========================
# NOTIFICATION SYSTEM
# =========================
message = f"Compliance Alert! Risk Level: {risk_level}. Missing clauses: {', '.join(missing_clauses)}"
cur.execute("INSERT INTO notifications VALUES (NULL, ?, ?)", (1, message))
conn.commit()

print("\nüìß Notification Sent:")
print(message)

# =========================
# FINAL REPORT
# =========================
print("\nüìä Compliance Risk Table:")
for row in cur.execute("SELECT * FROM compliance_risks"):
    print(row)

print("\nüì® Notification Log:")
for row in cur.execute("SELECT * FROM notifications"):
    print(row)

conn.close()
```
